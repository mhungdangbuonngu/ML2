```python
X = Valorant_players.to_numpy()
std_dev = np.array([132.094603882975,0.184304393260173,0.142124010550648])

X_scaled = np.empty_like(X)
for row in range(79):
  X_scaled[row,:] = X[row,:] / std_dev
X_scaled
#
```

```python
from sklearn.decomposition import PCA


pca = PCA(n_components=3)
pca.fit(X.transpose())

print(X)

print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
means = np.array([np.mean(X_scaled[:,0]),np.mean(X_scaled[:,1]),np.mean(X_scaled[:,2])])
means
```

```python
X_adj = np.empty_like(X_scaled)
for row in range(79):
  X_adj[row,:] = X_scaled[row,:] - means
X_adj
```

```python
cov_matrix = np.cov(X_adj.transpose())
cov_matrix
```

```python
eigenvalue, eigenvectors = np.linalg.eig(cov_matrix)
eigenvalue = np.real_if_close(eigenvalue,tol=1)
eigenvectors = np.real_if_close(eigenvectors,tol=1)
```

```python
eigenvalue, eigenvectors
```

```python
highest_eig_vector = np.argmax(eigenvalue)
PC1 = eigenvectors[highest_eig_vector,:]
second_highest_vector = np.argmax(np.delete(eigenvalue,0,0))
PC2 = eigenvectors[second_highest_vector + 1,:]
PC3 = eigenvectors[2,:]
PC1.shape, PC2.shape , PC3.shape
```

```python
V = np.array([PC1.transpose(),PC2.transpose()])
# V = np.array([PC1, PC2, PC3])
V
```

```python
V.shape , X_scaled.shape, X_adj.shape
```

```python
new_space = V.transpose() @ X_adj.transpose() @ X_scaled.transpose()
new_space.shape
```
